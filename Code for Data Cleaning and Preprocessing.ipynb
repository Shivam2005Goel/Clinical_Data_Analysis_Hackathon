## 1. Combine all the 9 Related Files present in  different study folders

### -> The below code will create 9 intermediate study files that consist of same type of data present in different studies

```python
import os
import pandas as pd

# Path to main QC folder
BASE_PATH = r"D:\Clinical-Data-Monitoring-System\QC Anonymized Study Files"

# All dataset keywords mapped to output names
DATASETS = {
    "Compiled_EDRR": "Combined_Compiled_EDRR_updated.xlsx",
    "CPID_EDC_Metrics": "Combined_CPID_EDC_Metrics.xlsx",
    "eSAE Dashboard": "Combined_eSAE_Dashboard.xlsx",
    "GlobalCodingReport_MedDRA": "Combined_GlobalCodingReport_MedDRA.xlsx",
    "GlobalCodingReport_WHODD": "Combined_GlobalCodingReport_WHODD.xlsx",
    "Inactivated Forms": "Combined_Inactivated_Forms.xlsx",
    "Missing_Lab_Name": "Combined_Missing_Lab_Ranges.xlsx",
    "Missing_Pages_Report": "Combined_Missing_Pages.xlsx",
    "Visit Projection Tracker": "Combined_Visit_Projection.xlsx"
}

# Function to combine all files for one dataset type
def combine_dataset(keyword, output_name):
    combined = []

    # Loop through each study folder
    for folder in os.listdir(BASE_PATH):
        folder_path = os.path.join(BASE_PATH, folder)

        if os.path.isdir(folder_path):
            for file in os.listdir(folder_path):

                if keyword in file and file.endswith((".xlsx", ".xls")):
                    full_path = os.path.join(folder_path, file)
                    print(f"Reading: {full_path}")

                    try:
                        df = pd.read_excel(full_path)

                        # Extract study number
                        study_name = folder.split("_")[0].strip()
                        df["Study"] = study_name

                        combined.append(df)

                    except Exception as e:
                        print(f"Error reading {full_path}: {e}")

    # Save combined output
    if combined:
        final_df = pd.concat(combined, ignore_index=True)
        output_path = os.path.join(BASE_PATH, output_name)
        final_df.to_excel(output_path, index=False)
        print(f"\nSaved combined file: {output_path}\n")
    else:
        print(f"\nNo files found for keyword: {keyword}\n")


# ---- MAIN AUTOMATION LOOP ----
for keyword, output_file in DATASETS.items():
    print("\n==============================")
    print(f"Combining dataset for keyword: {keyword}")
    print("==============================\n")
    combine_dataset(keyword, output_file)

print("\nðŸŽ‰ ALL DATASETS COMBINED SUCCESSFULLY!")
```

## 2. Generate the 3 final Output files for Analysis, DQI Calculation and Predictions from the dataset 


### -> The below code will generate the 3 Output Files 
###    a) Output_Patient_Level_Unified.xlsx
###    b) Output_Site_Level_Summary.xlsx
###    c)Output_High_Risk_Sites.xlsx
### -> It will calculate various other insights related to the output files generated

```python
import pandas as pd
import numpy as np
from datetime import datetime

# ========================================
# CLINICAL TRIAL DATA INTEGRATION SCRIPT
# ========================================

# 1. LOAD ALL DATA FILES
# ========================================
edc_metrics = pd.read_excel('./QC Anonymized Study Files/Combined_CPID_EDC_Metrics.xlsx')
edrr = pd.read_excel('./QC Anonymized Study Files/Combined_Compiled_EDRR_updated.xlsx')
esae = pd.read_excel('./QC Anonymized Study Files/Combined_eSAE_Dashboard.xlsx')
meddra = pd.read_excel('./QC Anonymized Study Files/Combined_GlobalCodingReport_MedDRA.xlsx')
whodd = pd.read_excel('./QC Anonymized Study Files/Combined_GlobalCodingReport_WHODD.xlsx')
inactivated = pd.read_excel('./QC Anonymized Study Files/Combined_Inactivated_Forms.xlsx')
missing_lab = pd.read_excel('./QC Anonymized Study Files/Combined_Missing_Lab_Ranges.xlsx')
missing_pages = pd.read_excel('./QC Anonymized Study Files/Combined_Missing_Pages.xlsx')
visit_proj = pd.read_excel('./QC Anonymized Study Files/Combined_Visit_Projection.xlsx')

print("Original EDC Metrics Columns:", edc_metrics.columns.tolist()[:15])

# 2. STANDARDIZE COLUMN NAMES
# ========================================
# Rename specific columns without creating duplicates
edc_metrics = edc_metrics.rename(columns={
    'Project Name': 'Study',
    'Site ID': 'Site_ID',
    'Subject ID': 'Subject_ID'
})

# Remove any duplicate columns (keep first)
edc_metrics = edc_metrics.loc[:, ~edc_metrics.columns.duplicated()]

print("Cleaned EDC Metrics Columns:", edc_metrics.columns.tolist()[:10])

# 3. BUILD BASE PATIENT TABLE
# ========================================
base_cols = ['Study', 'Region', 'Country', 'Site_ID', 'Subject_ID']
available_cols = [c for c in base_cols if c in edc_metrics.columns]

patient_base = edc_metrics[available_cols].drop_duplicates()
patient_base = patient_base.loc[:, ~patient_base.columns.duplicated()]
print(f"\nBase patient records: {len(patient_base)}")
print("Patient base columns:", patient_base.columns.tolist())

# 4. CALCULATE DERIVED METRICS
# ========================================

# 4.1 Missing Pages Count per Subject
missing_pages_count = missing_pages.groupby(
    ['Study', 'SiteNumber', 'SubjectName']
).size().reset_index(name='missing_pages_count')
missing_pages_count.columns = ['Study', 'Site_ID', 'Subject_ID', 'missing_pages_count']
print(f"Missing pages records: {len(missing_pages_count)}")

# 4.2 Open Issues Count (from EDRR)
open_issues = edrr.groupby(['Study', 'Subject']).agg({
    'Total Open issue Count per subject': 'sum'
}).reset_index()
open_issues.columns = ['Study', 'Subject_ID', 'total_open_issues']
print(f"Open issues records: {len(open_issues)}")

# 4.3 Uncoded Terms Count (MedDRA)
uncoded_meddra = meddra[meddra['Require Coding'] == 'Yes'].groupby(
    ['Study', 'Subject']
).size().reset_index(name='uncoded_meddra_terms')
uncoded_meddra.columns = ['Study', 'Subject_ID', 'uncoded_meddra_terms']
print(f"Uncoded MedDRA records: {len(uncoded_meddra)}")

# 4.4 Uncoded Terms Count (WHODD)
uncoded_whodd = whodd[whodd['Require Coding'] == 'Yes'].groupby(
    ['Study', 'Subject']
).size().reset_index(name='uncoded_whodd_terms')
uncoded_whodd.columns = ['Study', 'Subject_ID', 'uncoded_whodd_terms']
print(f"Uncoded WHODD records: {len(uncoded_whodd)}")

# 4.5 Missing Lab Count
lab_issues = missing_lab.groupby(
    ['Study', 'Subject', 'Site number']
).size().reset_index(name='missing_lab_count')
lab_issues.columns = ['Study', 'Subject_ID', 'Site_ID', 'missing_lab_count']
print(f"Lab issues records: {len(lab_issues)}")

# 5. BUILD UNIFIED PATIENT-LEVEL VIEW
# ========================================
unified_df = patient_base.copy()

print(f"\nBefore merges - unified_df columns: {unified_df.columns.tolist()}")

# Merge missing pages
unified_df = unified_df.merge(
    missing_pages_count.groupby(['Study', 'Subject_ID'])['missing_pages_count'].sum().reset_index(),
    on=['Study', 'Subject_ID'],
    how='left'
)

# Merge open issues
unified_df = unified_df.merge(
    open_issues,
    on=['Study', 'Subject_ID'],
    how='left'
)

# Merge uncoded MedDRA
unified_df = unified_df.merge(
    uncoded_meddra,
    on=['Study', 'Subject_ID'],
    how='left'
)

# Merge uncoded WHODD
unified_df = unified_df.merge(
    uncoded_whodd,
    on=['Study', 'Subject_ID'],
    how='left'
)

# Merge lab issues
unified_df = unified_df.merge(
    lab_issues.groupby(['Study', 'Subject_ID'])['missing_lab_count'].sum().reset_index(),
    on=['Study', 'Subject_ID'],
    how='left'
)

# 6. ENSURE ALL METRIC COLUMNS EXIST AND FILL NA
# ========================================
metric_columns = ['missing_pages_count', 'total_open_issues', 'uncoded_meddra_terms',
                  'uncoded_whodd_terms', 'missing_lab_count']

for col in metric_columns:
    if col not in unified_df.columns:
        unified_df[col] = 0
    else:
        unified_df[col] = unified_df[col].fillna(0).astype(int)

print(f"\nUnified DataFrame shape: {unified_df.shape}")
print("Final columns:", unified_df.columns.tolist())

# 7. CALCULATE CLEAN PATIENT STATUS
# ========================================
def calculate_clean_status(row):
    if (row['missing_pages_count'] == 0 and
        row['total_open_issues'] == 0 and
        row['uncoded_meddra_terms'] == 0 and
        row['uncoded_whodd_terms'] == 0):
        return 'Clean'
    return 'Not Clean'

unified_df['Clean_Patient_Status'] = unified_df.apply(calculate_clean_status, axis=1)

# 8. CALCULATE DATA QUALITY INDEX (DQI)
# ========================================
def calculate_dqi(row):
    weights = {'missing_pages': 0.30, 'open_issues': 0.30, 'uncoded_terms': 0.20, 'lab_issues': 0.20}

    missing_score = min(row['missing_pages_count'], 10) / 10
    issues_score = min(row['total_open_issues'], 10) / 10
    uncoded_score = min(row['uncoded_meddra_terms'] + row['uncoded_whodd_terms'], 10) / 10
    lab_score = min(row['missing_lab_count'], 10) / 10

    raw_score = (
        missing_score * weights['missing_pages'] +
        issues_score * weights['open_issues'] +
        uncoded_score * weights['uncoded_terms'] +
        lab_score * weights['lab_issues']
    )

    return round((1 - raw_score) * 100, 2)

unified_df['Data_Quality_Index'] = unified_df.apply(calculate_dqi, axis=1)

# 9. SITE-LEVEL AGGREGATIONS
# ========================================
groupby_cols = [c for c in ['Study', 'Region', 'Country', 'Site_ID'] if c in unified_df.columns]

site_summary = unified_df.groupby(groupby_cols).agg({
    'Subject_ID': 'count',
    'missing_pages_count': 'sum',
    'total_open_issues': 'sum',
    'uncoded_meddra_terms': 'sum',
    'uncoded_whodd_terms': 'sum',
    'missing_lab_count': 'sum',
    'Data_Quality_Index': 'mean',
    'Clean_Patient_Status': lambda x: (x == 'Clean').sum()
}).reset_index()

site_summary.columns = groupby_cols + [
    'Total_Subjects', 'Total_Missing_Pages', 'Total_Open_Issues',
    'Total_Uncoded_MedDRA', 'Total_Uncoded_WHODD', 'Total_Lab_Issues',
    'Avg_DQI', 'Clean_Patients_Count'
]

site_summary['Clean_Patient_Percentage'] = round(
    site_summary['Clean_Patients_Count'] / site_summary['Total_Subjects'] * 100, 2
)

# 10. IDENTIFY HIGH-RISK SITES (FIXED)
# ========================================
# Calculate Risk Score
site_summary['Risk_Score'] = (
    site_summary['Total_Missing_Pages'] * 0.25 +
    site_summary['Total_Open_Issues'] * 0.35 +
    (site_summary['Total_Uncoded_MedDRA'] + site_summary['Total_Uncoded_WHODD']) * 0.20 +
    (100 - site_summary['Avg_DQI']) * 0.20
)

# Print risk score distribution to understand data
print("\n" + "=" * 60)
print("RISK SCORE DISTRIBUTION:")
print("=" * 60)
print(site_summary['Risk_Score'].describe())
print(f"\nMax Risk Score: {site_summary['Risk_Score'].max()}")
print(f"Min Risk Score: {site_summary['Risk_Score'].min()}")

# Use percentile-based binning instead of fixed thresholds
percentiles = site_summary['Risk_Score'].quantile([0.25, 0.50, 0.75]).values
max_score = site_summary['Risk_Score'].max()
min_score = site_summary['Risk_Score'].min()

# Create unique bins by removing duplicates while maintaining order
unique_bins = [-float('inf')]
prev_val = -float('inf')

for p in percentiles:
    if p > prev_val:
        unique_bins.append(p)
        prev_val = p

if max_score > prev_val:
    unique_bins.append(float('inf'))
elif unique_bins[-1] != float('inf'):
    unique_bins.append(float('inf'))

# Create labels based on number of unique bins
num_categories = len(unique_bins) - 1
if num_categories == 4:
    labels = ['Low', 'Medium', 'High', 'Critical']
elif num_categories == 3:
    labels = ['Low', 'Medium', 'High']
elif num_categories == 2:
    labels = ['Low', 'High']
else:
    labels = ['Low']

# Apply binning only if we have more than one unique value
if max_score > min_score and len(unique_bins) > 2:
    site_summary['Risk_Level'] = pd.cut(
        site_summary['Risk_Score'],
        bins=unique_bins,
        labels=labels,
        duplicates='drop'
    )
else:
    # All sites have same or very similar risk scores
    site_summary['Risk_Level'] = 'Low'

# 11. EXPORT RESULTS
# ========================================
unified_df.to_excel('Output_Patient_Level_Unified.xlsx', index=False)
site_summary.to_excel('Output_Site_Level_Summary.xlsx', index=False)

# Export high-risk sites (top 25% or those with issues)
high_risk_sites = site_summary[
    (site_summary['Risk_Level'].isin(['High', 'Critical'])) |
    (site_summary['Risk_Score'] > site_summary['Risk_Score'].quantile(0.75))
].sort_values('Risk_Score', ascending=False)

# If still empty, export top 10 sites by risk score
if len(high_risk_sites) == 0:
    print("\nNo sites meet High/Critical criteria. Exporting top 10 sites by Risk Score instead.")
    high_risk_sites = site_summary.nlargest(min(10, len(site_summary)), 'Risk_Score')

high_risk_sites.to_excel('Output_High_Risk_Sites.xlsx', index=False)

# 12. SUMMARY STATISTICS
# ========================================
print("\n" + "=" * 60)
print("CLINICAL TRIAL DATA QUALITY SUMMARY")
print("=" * 60)
print(f"\nTotal Patients Analyzed: {len(unified_df)}")
print(f"Clean Patients: {(unified_df['Clean_Patient_Status'] == 'Clean').sum()}")
print(f"Clean Patient Rate: {round((unified_df['Clean_Patient_Status'] == 'Clean').mean() * 100, 2)}%")
print(f"\nTotal Sites: {len(site_summary)}")
print(f"High/Critical Risk Sites: {len(site_summary[site_summary['Risk_Level'].isin(['High', 'Critical'])])}")
print(f"Sites in High-Risk Report: {len(high_risk_sites)}")
print(f"\nAverage Data Quality Index: {round(unified_df['Data_Quality_Index'].mean(), 2)}")
print(f"Total Open Issues: {int(unified_df['total_open_issues'].sum())}")
print(f"Total Missing Pages: {int(unified_df['missing_pages_count'].sum())}")

# Risk level distribution
print("\n" + "=" * 60)
print("RISK LEVEL DISTRIBUTION:")
print("=" * 60)
print(site_summary['Risk_Level'].value_counts().sort_index())

print("\n" + "=" * 60)
print("OUTPUT FILES GENERATED:")
print("=" * 60)
print("1. Output_Patient_Level_Unified.xlsx")
print("2. Output_Site_Level_Summary.xlsx")
print(f"3. Output_High_Risk_Sites.xlsx ({len(high_risk_sites)} sites)")

# 13. DISPLAY TOP RESULTS
# ========================================
print("\n" + "=" * 60)
print("TOP 10 HIGH-RISK SITES:")
print("=" * 60)
top_10_cols = [c for c in ['Study', 'Site_ID', 'Total_Subjects', 'Risk_Score', 'Risk_Level', 'Avg_DQI'] if c in site_summary.columns]
print(site_summary.nlargest(10, 'Risk_Score')[top_10_cols])

# 14. ADDITIONAL DIAGNOSTICS
# ========================================
print("\n" + "=" * 60)
print("SITES WITH MOST ISSUES:")
print("=" * 60)
issue_cols = [c for c in ['Study', 'Site_ID', 'Total_Open_Issues', 'Total_Missing_Pages',
                          'Total_Uncoded_MedDRA', 'Total_Uncoded_WHODD'] if c in site_summary.columns]
print(site_summary.nlargest(5, 'Total_Open_Issues')[issue_cols])
```
